```python
import pandas as pd
import numpy as np
```

```python
u = pd.Series([9, 3, 5, 6, 8])
print(u)
```

```python
data = np.array([1, 3, 5, 7, 9])
v = pd.Series(data)
print(v)
```

```python
data = np.array(['a', 'b', 'c', 'd', 'e'])
v = pd.Series(data, index=[100, 101, 102, 103, 104])
print(v)
```

```python
dados = {'a': 0., 'b': 1., 'c': 2.}
v = pd.Series(dados)
print(v)
```

```python
idades = {'Ana': 25, 'Pedro': 30, 'Maria': 35}
i = pd.Series(idades)
print(i)
```

```python
data = {'a' : 0., 'b' : 1., 'c' : 2.}
v = pd.Series(data, index=['b', 'c', 'd', 'a'])
print(v)
```

```python
dados = pd.Series(5, index=[0, 1, 2, 3, 4, 5])
dados
```

```python
dados = pd.Series([1, 3, 5, 7, 9], index=['a', 'b', 'c', 'd', 'e'])
print(dados)

print(f"O primeiro elemento é:", dados[0])
print(f"O terceiro elemento é {dados[2]}")
```

```python
dados = pd.Series([1, 3, 5, 7, 9, 11, 13, 15], index=['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h'])

#buscando os três primeiros elementos
print("Buscando os três primeiros elementos")
print(dados[0:3])

print('Buscando os elementos de 3 a 6')
print(dados[2:6])

print('Buscando os três últimos elementos')
print(dados[-3:])
```
