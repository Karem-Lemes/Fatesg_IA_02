```python
import numpy as np
#teste01
```

```python
print("--- Exercício 1 ---")

arr1 = np.arange(1, 11)
print(f"Array original: {arr1}")


arr1[5:9] = 0
print(f"Array modificado: {arr1}\n")
```

```python
print("--- Exercício 2 ---")

arr2 = np.arange(1, 7).reshape(3, 2)
print(f"Matriz:\n{arr2}")

print(f"Shape da matriz: {arr2.shape}")

print(f"2ª linha: {arr2[1]}\n")
```

```python
print("--- Exercício 3 ---")

arr3 = np.array([[10, 20],
                 [30, 40],
                 [50, 60]])
print(f"Matriz:\n{arr3}")

print(f"2ª coluna: {arr3[:, 1]}\n")
```

```python
print("--- Exercício 4 ---")

arr4 = np.arange(1, 21).reshape(4, 5)
print(f"Matriz:\n{arr4}")

print(f"3ª linha: {arr4[2]}\n")
```

```python
print("--- Exercício 5 ---")

arr5 = np.arange(101, 121).reshape(4, 5)
print(f"Matriz:\n{arr5}")

print("Elementos (linhas 1 e 2 | colunas 2 e 3):")
print(f"{arr5[0:2, 1:3]}\n")
```

```python
print("--- Exercício 6 ---")

arr6 = np.random.randint(1, 100, size=(4, 5))
print(f"Matriz:\n{arr6}")

print("Elementos (linhas 2 e 3 | colunas 1 a 3):")
print(f"{arr6[1:3, 0:3]}\n")
```
