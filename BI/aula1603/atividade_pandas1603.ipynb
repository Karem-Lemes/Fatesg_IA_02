```python
import pandas as pd
import numpy as np
```

```python
dados = {
    'Frutas': ['Maçã', 'Banana', 'Morango'],
    'Quantidade': [10, 20, 30]
}

df = pd.DataFrame(dados)
print(df)
```

```python
data = [{'a': 1, 'b': 2}, {'a': 5, 'b': 10, 'c': 20}]
df = pd.DataFrame(data)
print(df)
```

```python
serie1 = pd.Series([1, 2], index=['a', 'b'])
serie2 = pd.Series([3, 4], index=['b', 'c'])

dados = {'col1': serie1, 'col2': serie2}
df = pd.DataFrame(dados)

print(df)
```

```python
df = pd.DataFrame({
    'one': [10, 20, 30],
    'two': [5, 5, 5]
})

df['three'] = df['one'] + df['two']

print(df)
```

```python
df = pd.DataFrame({
    'A': [1, 2, 3],
    'B': [4, 5, 6],
    'C': [7, 8, 9]
})

del df['A']

coluna_removida = df.pop('B')

print("DataFrame resultante:")
print(df)
print("\nConteúdo retornado pelo pop:")
print(coluna_removida)
```

```python
df = pd.DataFrame({
    'Valor': [10, 20, 30]
}, index=['a', 'b', 'c'])

linha_b = df.loc['b']

print(linha_b)
```

```python
df = pd.DataFrame({
    'Nome': ['Ana', 'Beto', 'Caio', 'Duda'],
    'Idade': [20, 25, 30, 35]
})

# iloc seleciona pela posição inteira (começando em 0)
# A posição 2 corresponde à terceira linha
terceira_linha = df.iloc[2]

print(terceira_linha)
```

```python
df_vendas = pd.DataFrame({
    'Produto': ['A', 'B', 'C', 'D', 'E'],
    'Vendas': [100, 150, 200, 250, 300]
})

print(df_vendas.head(3))
```

```python
df = pd.DataFrame({
    'A': range(7),
    'B': range(7),
    'C': range(7)
})

print(df.shape)
```

```python
df = pd.DataFrame({
    'A': [1, 2],
    'B': [3, 4]
})

print("Transposta (df.T):")
print(df.T)

print("\nTipos de dados (df.dtypes):")
print(df.dtypes)

print("\nEstá vazio? (df.empty):")
print(df.empty)
```
