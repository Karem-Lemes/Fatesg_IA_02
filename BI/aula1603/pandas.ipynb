```python
import pandas as pd
import numpy as np
```

```python
data = {'Name': ['João', 'Maria', 'Pedro', 'Ana'], 'Age': [25, 30, 35, 20   ], 'City': ['São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Curitiba']}
df = pd.DataFrame(data)
df
```

```python
df = pd.DataFrame(data, index = ['rank1', 'rank2', 'rank3', 'rank4'])
df
```

```python
data = [{'a': 1, 'b': 2}, {'a': 3, 'b': 4}, {'a': 5, 'b': 6}, {'a': 7, 'b': 8}]
df = pd.DataFrame(data)
df
```

```python
data = [{'a': 1, 'b': 2}, {'a': 3, 'b': 4}, {'a': 5, 'b': 6}, {'a': 7, 'b': 8}]
df1 = pd.DataFrame(data, index = ['first', 'second', 'third', 'fourth'], columns = ['a', 'b'])
df1

df2 = pd.DataFrame(data, index = ['first', 'second', 'third', 'fourth'], columns = ['a', 'b1'])

print(df1)
print(df2)
```

```python
dados = {'one': pd.Series([1.,2.,3.], index=['a','b','c']),
         'two': pd.Series([4.,5.,6.], index=['a','b','c'])}
df = pd.DataFrame(dados)
df
```

```python
print("Adicionando uma nova coluna passando com Series:")
df['three'] = pd.Series([10,20,30], index=['a','b','c'])
df

print("Adicionando uma nova coluna usando as colunas existentes no DataFrame:")
df['four'] = df['one'] + df['three']
df
```

```python
d = {'one': pd.Series([1,2,3], index=['a','b','c']),
    'two': pd.Series([1,2,3,4], index=['a','b','c','d']),
    'three': pd.Series([10,20,30], index=['a','b','c'])}

df = pd.DataFrame(d)
print("O dataframe é:")
print(df)
```

```python
print("Excluindo a primeira coluna usando a função DEL:")
del df['one']
print(df)
```

```python
print("Excluindo a primeira linha usando a função POP:")
df.pop('two')
print(df)
```

```python
d = {'one': pd.Series([1,2,3], index=['a','b','c']),
    'two': pd.Series([1,2,3,4], index=['a','b','c','d'])}

df = pd.DataFrame(d)
print(df)
print(28*"#")
print(df.loc['b'])
```

```python
d = {'one': pd.Series([1,2,3], index=['a','b','c']),
    'two': pd.Series([1,2,3,4], index=['a','b','c','d'])}

df = pd.DataFrame(d)
print(df.iloc[2])
```

```python
d = {'one': pd.Series([1,2,3], index=['a','b','c']),
    'two': pd.Series([1,2,3,4], index=['a','b','c','d'])}

df = pd.DataFrame(d)
print(df[2:4])
```
